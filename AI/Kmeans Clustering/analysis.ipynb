```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Load text file with pasted data
data = np.loadtxt("results/results-iris.txt")

## Calculate the mean -percentage- correct
## We used 10-fold random subsampling validation (10.0)
## Apply the np.mean() function across the rows (axis 0, one mean per column)
means = np.apply_along_axis(np.mean,0,data)*100.0/10.0


## Standard error is the standard deviation (np.std) / sqrt(N-1)
## Alternatively, we just use scipy.stats.sem() [sps.sem()]
stderr = np.apply_along_axis(sps.sem,0,data)*100.0/10.0
## Make the plot bigger to get a better resolution
plt.figure(figsize=(15,5));
## Use an error bar plot...
plt.errorbar(np.array(range(len(means)))+1.0,means,yerr=stderr);
## Fix the axis labels
plt.xlabel('Number of Clusters (K)');
plt.ylabel('Percentage Correct (%) [+- Std. Err.]');

#Note that the iris data set can be sufficiently represented using only 7 clusters (maybe 10 if you want to be a little more conservative).
#That is, the percentage correct for K=7 is -statistically- indistingushable from using higher K=8,9,10,... and suggests that a decent non-parametric model of the data set consists of only 7 vectors.
#This means that remembering only 7 vectors instead of 150 results in a compression ratio of almost 22:1 (or less than 5% of the original data size).
```

```python
## Same calculation on the cancer data set
data = np.loadtxt("results/results-cancer.txt")

means = np.apply_along_axis(np.mean,0,data)*100.0/10.0
stderr = np.apply_along_axis(sps.sem,0,data)*100.0/10.0
plt.figure(figsize=(15,5));
plt.errorbar(np.array(range(len(means)))+1.0,means,yerr=stderr);
plt.xlabel('Number of Clusters (K)');
plt.ylabel('Percentage Correct (%) [+- Std. Err.]');
```
