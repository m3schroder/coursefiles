# Problem 3

```python
x = sp.symbols('x')
y = 1 / (1 + sp.exp(-x))

dydx = y.diff(x)
x_vals = np.linspace(-5,5)
y_vals = np.array([dydx.evalf(subs=dict(x=x_val)) for x_val in x_vals])

plt.plot(x_vals,y_vals)
plt.title('$y=%s$'%(sp.latex(dydx)))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
